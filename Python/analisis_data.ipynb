## Agrupar por fecha

```python
# Convertir la columna 'Date' a formato datetime si no está en ese formato
launches_no_nulos['Date'] = pd.to_datetime(launches_no_nulos['Date'], errors='coerce')

# Filtrar las fechas del año 2000
launches_no_nulos_2000 = launches_no_nulos[launches_no_nulos['Date'].dt.year == 2018]

# Mostrar el DataFrame filtrado
launches_no_nulos_2000
```

### ¿Qué organización lanzó más misiones espaciales en un año determinado?
  año 2018



```python
# Agrupar por título y calcular el promedio y la mediana del salario para cada título
df_grouped = launches_no_nulos_2000.groupby('Organization')['Date'].agg(['count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['Organization',  'Organization_count']

# Mostrar el DataFrame resultante
df_grouped
```

```python
# Ordenar el DataFrame por la columna 'Organization_count' de manera descendente
df_grouped_sorted = df_grouped.sort_values(by='Organization_count', ascending=False)

# Mostrar el DataFrame ordenado
print("Títulos ordenados de mayor a menor por número de lanzamientos:")
print(df_grouped_sorted)
```

```python
# Obtener el título con el mayor número de lanzamientos
titulo_max_lanzamientos = df_grouped.loc[df_grouped['Organization_count'].idxmax(), 'Organization']

print("Título con el mayor número de lanzamientos:", titulo_max_lanzamientos)
```

### ¿Cuántos lanzamientos se realizaron por año? 

```python
launches_no_nulos
```

```python
# Extraer el año de la fecha y crear una nueva columna 'Year'
launches_no_nulos['Year'] = launches_no_nulos['Date'].dt.year

# Agrupar por año y contar el número de lanzamientos en cada año
lanzamientos_por_año = launches_no_nulos.groupby('Year').size()

# Mostrar los lanzamientos por año
lanzamientos_por_año
```

### Gráfica

```python
# Graficar los lanzamientos por año con un tamaño de gráfico más grande
plt.figure(figsize=(12, 8))
lanzamientos_por_año.plot(kind='bar', color='skyblue')
plt.title('Lanzamientos por año')
plt.xlabel('Año')
plt.ylabel('Número de lanzamientos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Costo a raiz del tiempo

```python
# Calcular el costo promedio de las misiones para cada año
costo_promedio_por_año = launches_no_nulos.groupby(launches_no_nulos['Date'].dt.year)['Price'].mean()

# Graficar el costo promedio de las misiones a lo largo del tiempo
plt.figure(figsize=(12, 6))
costo_promedio_por_año.plot(marker='o', linestyle='-', color='b')
plt.title('Costo promedio de las misiones espaciales por año (1954-2020)')
plt.xlabel('Año')
plt.ylabel('Costo promedio de la misión')
plt.grid(True)
plt.xticks(range(1954, 2021), rotation=45)
plt.tight_layout()
plt.show()
```

### ¿Qué meses son los más populares para los lanzamientos? 


```python
# Extraer el mes de la fecha y crear una nueva columna 'Month'
launches_for_month = launches_no_nulos
launches_for_month['Month'] = launches_no_nulos['Date'].dt.month

# Contar el número de lanzamientos en cada mes
lanzamientos_por_mes = launches_for_month['Month'].value_counts()

# Encontrar los meses con el mayor número de lanzamientos
meses_populares = lanzamientos_por_mes[lanzamientos_por_mes == lanzamientos_por_mes.max()]

meses_populares
```

```python
import matplotlib.pyplot as plt

# Ordenar los meses en orden cronológico
meses_ordenados = lanzamientos_por_mes.sort_index()

# Graficar la cantidad de lanzamientos por mes
plt.figure(figsize=(10, 6))
meses_ordenados.plot(kind='bar', color='skyblue')
plt.title('Cantidad de lanzamientos por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de lanzamientos')
plt.xticks(range(0, 12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
